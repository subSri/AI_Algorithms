```python
#checking if coming state already in openlist

def isdictVisited(state,vis):
  flag1=0
  for i in vis:
    if (state==(np.fromstring(i,dtype=state.dtype).reshape(3,3))).all()==True:
      flag1=1
      break
  if flag1==1:
    return True
  else:
    return False
```

```python
#checking if current state in closedList

def isVisited(state,vis):
  flag=0
  for i in vis:
    if (state==i).all()==True:
      flag=1
      break
  if flag==1:
    return True
  else:
    return False
```

```python
#Checking if current state is Goal

def isGoal(state):
  return (goal==state).all()
```

```python
#if current state already in openList then Update its f(N) in the Hash Table

def update_in_Hash(myhash,state,state_fn):
    state_bytes = state.tostring()
    if myhash[state_bytes]>state_fn:
        myhash[state_bytes]=state_fn
```

```python
#Get Manhattan Heuristic Value of a particular state

def get_heuristics(state):
    coord_list=[]
    state_list=[]
    for i in range(3):
        for j in range(3):
            coord_list.append([i,j])
#     print(state)
    for i in range(8):
        k=list(np.where(state==i+1))
#         print(k)
        state_list.append([k[0][0],k[1][0]])
    heur = 0
    for i in range(8):
        heur = heur + abs(coord_list[i][0]-state_list[i][0]) + abs(coord_list[i][1]-state_list[i][1])
    return heur
        
```

```python
#Finds state with minimum f(N) and expands it to find other state and recursively calls the next states to add into the Hash
#find thier minimum everytime and Finally reach the goal state if possible.
#I am taking key of matrix in Hash as a string to save space

def findGoal(curr_state,gn,openlist,closedlist,opit):
  if isGoal(np.array(curr_state)):
    print("Congrats you reached the goal\n",curr_state)
    
    return True
  elif opit>2000:
    return False
  else:
    
    possib = [[0,-1],[1,0],[0,1],[-1,0]]
   
    solution = False
    least_state_bytes = min(openlist, key=openlist.get)
    least_state = (np.fromstring(least_state_bytes,dtype=curr_state.dtype)).reshape(3,3)
    print("Curr State :\n", least_state,"\n")
    del openlist[least_state_bytes]
    closedlist.append(least_state)
    myI,myJ=map(int,np.where(least_state==0))
    for i,j in possib:
      
      if (myI+i>=0 and myJ+j<3 and myI+i<3 and myJ+j>=0):
        
        next_state = deepcopy(least_state)
        temp = least_state[myI+i][myJ+j]
        next_state[myI+i][myJ+j] = 0
        next_state[myI][myJ] = temp
        if isdictVisited(next_state,openlist.keys()):
            update_in_Hash(openlist,next_state,get_heuristics(next_state)+gn)
        elif not isVisited(next_state,closedlist):
          next_state_bytes = next_state.tostring()
          openlist[next_state_bytes]=get_heuristics(next_state)+gn
        else:
            next_state = deepcopy(least_state)
    
    
    solution = findGoal(least_state,gn+1,openlist,closedlist,opit+1)
    return solution


```
